```python
import pandas as pd
import ast
import csv
```

```python
af_data_folder ="/Users/adamtsuchiyama/Documents/nss_capstone/data/"
```

```python
file = open(af_data_folder + 'audio_features_dic.txt', 'r')
contents = file.read()
audio_features_list = ast.literal_eval(contents)
file.close()
audio_features_list
```

```python
def create_song_csv(csv_name, headings_list, data_list):
    "creates a csv file from a list spotify data"
    with open(csv_name, 'w') as out_file:
        writer = csv.DictWriter(out_file, headings_list)
        writer.writeheader()
        for row in data_list:
            if row:
                writer.writerow(row)
        return;
```

```python
headings_af = ['spotify_id', 'tempo', 'time_signature', 'key', 'mode', 'loudness', 'energy', 'danceability', 'acousticness', 'instrumentalness', 'liveness', 'speechiness', 'valence']
```

```python
create_song_csv('audio_features.csv', headings_af, audio_features_list)
```
