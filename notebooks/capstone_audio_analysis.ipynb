```python
def create_song_csv(csv_name, headings_list, data_list):
    "creates a csv file from a list spotify data"
    with open(csv_name, 'w') as out_file:
        writer = csv.DictWriter(out_file, headings_list)
        writer.writeheader()
        for row in data_list:
            if row:
                writer.writerow(row)
        return;
```

```python
queen_rem_seg = pd.read_csv('/Users/adamtsuchiyama/Documents/nss_capstone/data/bites_the_dust_rem_seg.csv')
queen_rem_seg['start_rnd'] = round(queen_rem_seg['start'], 1)
queen_rem_seg['duration_rnd'] = round(queen_rem_seg['duration'], 1)
queen_rem_seg = queen_rem_seg[['loudness_start', 'loudness_max', 'pitches', 'timbre', 'start_rnd', 'duration_rnd']]
queen_rem_seg.columns = ['rem_' + str(col) for col in queen_rem_seg.columns]
queen_rem_seg = queen_rem_seg.rename(columns={'rem_start_rnd' : "start_rnd", 'rem_duration_rnd' : 'duration_rnd'})
queen_rem_seg.head()
```

```python
queen_seg_mrge = pd.merge(queen_seg, queen_rem_seg, on='start_rnd')
queen_seg_mrge = queen_seg_mrge[['start_rnd', 'loudness_start', 'rem_loudness_start', 'loudness_max', 'rem_loudness_max', 'pitches', 'rem_pitches', 'timbre', 'rem_timbre']]
queen_seg_mrge.head()
```

```python
x = 0
pitch_eucl_dist = []
while x <= len(queen_seg_mrge['pitches'])-1:
    pitch_val = json.loads(queen_seg_mrge.loc[x, 'pitches'])
    rem_val = json.loads(queen_seg_mrge.loc[x, 'rem_pitches'])
    pitch_val = np.array(pitch_val)
    rem_val = np.array(rem_val)
    eucl_dist = distance.euclidean(pitch_val, rem_val)
    pitch_eucl_dist.append(eucl_dist)
    x += 1
```

```python
pitch_eucl_dist
```

```python
queen_seg_mrge['pitch_eucl_dist'] = pitch_eucl_dist
queen_seg_mrge.head()
```

```python
x = 0
timbre_eucl_dist = []
while x <= len(queen_seg_mrge['timbre'])-1:
    timbre_val = json.loads(queen_seg_mrge.loc[x, 'timbre'])
    rem_val = json.loads(queen_seg_mrge.loc[x, 'rem_timbre'])
    timbre_val = np.array(timbre_val)
    rem_val = np.array(rem_val)
    eucl_dist = distance.euclidean(timbre_val, rem_val)
    timbre_eucl_dist.append(eucl_dist)
    x += 1
timbre_eucl_dist
```

```python
queen_seg_mrge['timbre_eucl_dist'] = timbre_eucl_dist
queen_seg_mrge.head()
```
