```python
import pandas as pd
import ast
import csv
```

```python
file = open("another_one_bites_the_dust_audio_analysis.txt", 'r')
contents = file.read()
dictionary = ast.literal_eval(contents)
file.close()
bites_dust_seg = dictionary['segments']
```

```python
headings = ['start', 'duration', 'confidence', 'loudness_start', 'loudness_max_time', 'loudness_max', 'loudness_end', 'pitches', 'timbre']
```

```python
def create_song_csv(csv_name, headings_list, data_list):
    "creates a csv file from a list spotify data"
    with open(csv_name, 'w') as out_file:
        writer = csv.DictWriter(out_file, headings_list)
        writer.writeheader()
        for row in data_list:
            if row:
                writer.writerow(row)
        return;
```

```python
create_song_csv('bites_the_dust_seg.csv', headings, bites_dust_seg)
```

```python
bites_the_dust_seg = pd.read_csv('bites_the_dust_seg.csv')
bites_the_dust_seg
```

```python
create_song_csv('bites_the_dust_rem_seg.csv', headings, bites_dust_rem_seg)
```
