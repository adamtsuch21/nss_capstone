```python
def create_song_csv(csv_name, headings_list, data_list):
    "creates a csv file from a list spotify data"
    with open(csv_name, 'w') as out_file:
        writer = csv.DictWriter(out_file, headings_list)
        writer.writeheader()
        for row in data_list:
            if row:
                writer.writerow(row)
        return;
```

```python
def txt_to_csv(path_list, headings):
    "creates csvs from a list of txt file paths"
    for path in path_list:
        file = open(path, 'r')
        contents = file.read()
        dictionary = ast.literal_eval(contents)
        file.close()
        data = dictionary['segments']
        file_name = path.split('/')
        file_name = file_name[7]
        file_name = file_name.split('.')
        file_name = file_name[0] + '.csv'
        create_song_csv(file_name, headings, data)
    return;
```

```python
def time_rnd(df, file_path):
    "rounds the time from a spotify df, and adds rem to col names (if remaster data)"
    df['start_rnd'] = round(df['start'], 1)
    df['duration_rnd'] = round(df['duration'], 1)
    return;
```

```python
queen_og = pd.read_csv(queen_fp + 'analysis.csv')
time_rnd(queen_og, 'queen_analysis.csv')
queen_og = queen_og.drop(columns=['confidence', 'loudness_max_time', 'loudness_end'])
queen_og.head()
```

```python
queen_rem = pd.read_csv(data_folder + 'rem_analysis.csv')
time_rnd(queen_rem, 'queen_rem_analysis.csv')
queen_rem = queen_rem.drop(columns=['confidence', 'loudness_max_time', 'loudness_end'])
queen_rem.columns = ['rem_' + str(col) for col in queen_rem.columns]
queen_rem.head()
```

```python
queen_mrge = pd.merge(queen_og, queen_rem, left_on=['start_rnd', 'duration_rnd'], right_on=['rem_start_rnd', 'rem_duration_rnd'])
queen_mrge = queen_mrge[['start_rnd', 'duration_rnd', 'loudness_start', 'rem_loudness_start', 'loudness_max', 'rem_loudness_max', 'pitches', 'rem_pitches', 'timbre', 'rem_timbre']]
queen_mrge.head()
```

```python
def spotify_eucl_dist(df, first_col, second_col, eucl_list):
    "calculates euclidean distance between two columns in a spotify df"
    x = 0
    while x <= len(df[first_col])-1:
        og_val = json.loads(df.loc[x, first_col])
        rem_val = json.loads(df.loc[x, second_col])
        og_val = np.array(og_val)
        rem_val = np.array(rem_val)
        eucl_dist = distance.euclidean(og_val, rem_val)
        eucl_list.append(eucl_dist)
        x += 1
    return;
```

```python
queen_pitch_eucl = []
spotify_eucl_dist(queen_mrge, 'pitches', 'rem_pitches', queen_pitch_eucl)
queen_pitch_eucl
```

```python
queen_timbre_eucl = []
spotify_eucl_dist(queen_mrge, 'timbre', 'rem_timbre', queen_timbre_eucl)
queen_timbre_eucl
```

```python
def spotify_new_cols(df, pitch_eucl_list, timbre_eucl_list):
    "creates new columns in a spotify df from pitch and timbre eucl dist lists, as well as loudness diff columns"
    df['pitch_eucl_dist'] = pitch_eucl_list
    df['timbre_eucl_dist'] = timbre_eucl_list
    df['loud_start_diff'] = df['rem_loudness_start'] - df['loudness_start']
    df['loud_max_diff'] = df['rem_loudness_max'] - df['loudness_max']
    return;
```

```python
spotify_new_cols(queen_mrge, queen_pitch_eucl, queen_timbre_eucl)
queen_mrge = queen_mrge[['start_rnd', 'duration_rnd', 'loudness_start', 'rem_loudness_start', 'loud_start_diff', 'loudness_max', 'rem_loudness_max', 'loud_max_diff', 'pitches', 'rem_pitches', 'pitch_eucl_dist', 'timbre', 'rem_timbre', 'timbre_eucl_dist']]
queen_mrge.head()
```
